```python
## de acuerdo con las instrucciones de la tarea importamos los datos
## de wiki usando pandas
df_dattoronto = pd.read_html('https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=1008658788')
df_dattoronto
```

```python
## verificar cuantas tablas hay en el conjunto de archivos en wiki
print(f'El contenido total de tablas es de -->: {len(df_dattoronto)}')
```

```python
## Creación del dataframe para datos de codigos postales
df_codigos_toronto= df_dattoronto[0]
df_codigos_toronto.head(20)
```

-------------------------------------
Puede existir más de un vecindario en un área de código postal. Por ejemplo, en la tabla de la página de Wikipedia, notará que M5A aparece dos veces y tiene dos vecindarios: Harbourfront y Regent Park. Estas dos filas se combinarán en una fila con los vecindarios separados con una coma como se muestra en la fila 11 en la tabla anterior
----------------------------------------

```python
## Eliminacion de datos incompletos para municipios sin asignacion 
df_codigos_toronto1 = df_codigos_toronto[df_codigos_toronto ['Borough']!= "Not assigned"]
df_codigos_toronto1 = df_codigos_toronto1.reset_index(drop=True)
df_codigos_toronto1.head(20)
```

```python
## verificar si se realizo bien la instrucción anterior
df_codigos_toronto1[df_codigos_toronto1['Neighbourhood'] == 'Not assigned']
```
