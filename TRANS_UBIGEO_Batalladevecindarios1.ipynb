```python
## MUESTRA LAS COLUMNAS DEL ARCHIVO ORIGINAL CON VALORES NULOS Y MUESTRA SI ES REPRESENTATIVO O NO SOBRE EL TOTAL DE REGISTRO
i = 0
ni = len(df_original.columns)
nombre_columna = df_original.columns
while i < ni:
  print("CANTIDAD DE DATOS NULOS EN COLUMNA ", i+1, " (", nombre_columna[i], ") ES DE :", sum(df_original[nombre_columna[i]].isnull()))
  val_col=sum(df_original[nombre_columna[i]].isnull())
  val1_col=round(float(val_col/Tot_filas),4)
  print("estos valores nulos representan-->", val1_col, "del total de registros")
  i = i +1
  if i == ni:
    break
```
