```python
# usamos el codigo del programa anterior
## de acuerdo con las instrucciones de la tarea importamos los datos
## de wiki usando pandas
df_dattoronto = pd.read_html('https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=1008658788')
df_codigos_toronto = df_dattoronto[0]
df_codigos_toronto = df_codigos_toronto[df_codigos_toronto['Borough']!= 'Not assigned']
df_codigos_toronto = df_codigos_toronto.reset_index(drop=True)
df_codigos_toronto.head(20)
```

```python
## ahora se indica cual es el dato a buscar
postal_code = df_codigos_toronto['Postal Code']
postal_code
```

```python
## ahora de acuerdo a la instrucción se tienen que determinar las coordenadas 
!wget -O toronto_coord.csv http://cocl.us/Geospatial_data 
df_toronto_coordenadas = pd.read_csv('toronto_coord.csv')
print(df_toronto_coordenadas.head())
print(df_toronto_coordenadas.shape)
```

```python
## miremos el contenido del panda df_toronto_coordnadas
df_toronto_coordenadas.head(20)
```

```python
## unificamos la informacion de los dos pandas con datos
df_toronto = pd.merge(df_codigos_toronto,df_toronto_coordenadas)
df_toronto.head(20)
```

```python
## verificamos el tamaño del nuevo panda con la union anterior
df_toronto.shape
```

```python
## segunda parte
## cargar el geopy 
## verifiqe si esta geopy cargado en conda sino instale manualmente
# conda install -c conda-forge geopy ** y le da Y
from geopy.geocoders import Nominatim
# carga folium
!pip install folium
import folium
```

```python
# Buscamos las coordenadas para realizar el mapa
address = 'Toronto, Ontario'
geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('Las coordenadas de '+address,' son {}, {}.'.format(latitude,longitude))
```

```python
# hacemos el  mapa de Toronto City
mapa_toronto = folium.Map(location=[latitude, longitude], zoom_start = 10)
mapa_toronto
```

```python
# ahora ponemos los marcadores de los municipios y vecindarios
for lat,lng, borough, neighborhood in zip(df_toronto['Latitude'],df_toronto['Longitude'],df_toronto['Borough'],df_toronto['Neighbourhood']):
    label= 'Neighborhood {}, Borough {}'.format(neighborhood,borough)
    label= folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat,lng],
        radius=5,
        popup= label,
        color= 'red',
        fill=True,
        fill_color='yellow',
        fill_opacity=0.7,
        parse_html=False).add_to(mapa_toronto)

mapa_toronto
```

```python
# revisa los municipios de Toronto y contamos las repeticiones , que nos indican cuantos vecindarios tiene cada uno
df_toronto_borough = pd.value_counts(df_toronto['Borough'])
df_toronto_borough = pd.DataFrame(df_toronto_borough).reset_index()
df_toronto_borough = df_toronto_borough.rename(columns={'index':'Borough','Borough':'Number of Neighborhood'})
df_toronto_borough
```
